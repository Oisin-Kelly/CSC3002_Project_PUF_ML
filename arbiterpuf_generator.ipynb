```python
from models.ArbiterPUF import ArbiterPUF
from helpers import get_XY, save_to_memmap
import os
```

Creating folders

```python
os.makedirs("crps/arbiter_puf/0noise", exist_ok=True)
os.makedirs("crps/arbiter_puf/5noise", exist_ok=True)
os.makedirs("crps/arbiter_puf/5noise_majority", exist_ok=True)
os.makedirs("crps/arbiter_puf/10noise", exist_ok=True)
os.makedirs("crps/arbiter_puf/10noise_majority", exist_ok=True)
os.makedirs("crps/arbiter_puf/20noise", exist_ok=True)
os.makedirs("crps/arbiter_puf/20noise_majority", exist_ok=True)
```

Generating for 64 bit APUFs (0% noise)

```python
arbiter_puf = ArbiterPUF(64, noise=0)

responses = arbiter_puf.generate_challenges_reponses(100_000)

challenges, responses = get_XY(responses)

save_to_memmap(challenges, "crps/arbiter_puf/0noise/APUF_64bit_chal_100000.memmap")
save_to_memmap(responses, "crps/arbiter_puf/0noise/APUF_64bit_resp_100000.memmap")
```
