```python
import numpy as np
```

```python
def calculate_uniformity(responses):
    C = len(responses)

    if C == 0:
        return 0.0
    sum_ones = np.sum(responses)  
    U = (sum_ones / C) * 100
    
    return U
```

```python
responses = np.memmap("crps/arbiter_puf/0noise/APUF_64bit_resp_100000.memmap", shape=(100_000))

print("64bit uniformity:", calculate_uniformity(responses))

responses = np.memmap("crps/arbiter_puf/0noise/APUF_64bit_resp_100000.memmap", shape=(100_000))

print("128bit uniformity:", calculate_uniformity(responses))
```
